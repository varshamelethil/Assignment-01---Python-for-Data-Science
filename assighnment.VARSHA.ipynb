The Iris Dataset contains four features(Sepal length,Sepal width,Petal length and Petal width) of 50 samples of three species of Iris(Iris Setosa,Iris Versicolour and Iris virginica)



sepal length (SL) in cm, sepal Width (SW) in cm, petal length (PL) in cm, petal width (PW)in cm

### 5. Perform meaningful visualizations using the dataset 

#### Scatter Plots 

```python
plot=dataset[dataset.Classification=='Iris-setosa'].plot.scatter(x='PL',y='PW',color='red',label='Iris Setosa')
dataset[dataset.Classification=='Iris-versicolor'].plot.scatter(x='PL',y='PW',color='blue',label='Iris Versicolor',ax=plot)
dataset[dataset.Classification=='Iris-virginica'].plot.scatter(x='PL',y='PW',color='green',label='Iris Virginica',ax=plot)
plt.xlabel('PL')
plt.ylabel('PW')
plt.title('Petal Length vs Petal Width')
plt.show()
```

Setosa species have the smallest petal length as well as petal width.Versicolor species have average petal length and petal width.Virginica species have the highest petal length as well as petal width


```python
plot=dataset[dataset.Classification=='Iris-setosa'].plot.scatter(x='SL',y='SW',color='red',label='Iris Setosa')
dataset[dataset.Classification=='Iris-versicolor'].plot.scatter(x='SL',y='SW',color='blue',label='Iris Versicolor',ax=plot)
dataset[dataset.Classification=='Iris-virginica'].plot.scatter(x='SL',y='SW',color='green',label='Iris Virginica',ax=plot)
plt.xlabel('SL')
plt.ylabel('SW')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

Iris Setosa species has smaller sepal length but higher width.Versicolor lies in almost middle for length as well as width.Virginica has larger sepal lengths and smaller sepal widths

```python
plot=dataset[dataset.Classification=='Iris-setosa'].plot.scatter(x='PL',y='SL',color='red',label='Iris Setosa')
dataset[dataset.Classification=='Iris-versicolor'].plot.scatter(x='PL',y='SL',color='blue',label='Iris Versicolor',ax=plot)
dataset[dataset.Classification=='Iris-virginica'].plot.scatter(x='PL',y='SL',color='green',label='Iris Virginica',ax=plot)
plt.xlabel('PL')
plt.ylabel('SL')
plt.title('Petal Length vs Sepal Length')
plt.show()
```

Iris Setosa species has smaller Petal length.Versicolor lies in almost middle for length as well as width.Virginica has larger petal length.

#### Box plot 

```python
sns.boxplot(x='Classification', y='PL', data=dataset)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16,9))
sns.boxplot( y='PW', x= 'Classification', data=dataset, orient='v' , ax=axes[0, 0])
sns.boxplot( y='PL', x= 'Classification', data=dataset, orient='v' , ax=axes[0, 1])
sns.boxplot( y='SL', x= 'Classification', data=dataset, orient='v' , ax=axes[1, 0])
sns.boxplot( y='SW', x= 'Classification', data=dataset, orient='v' , ax=axes[1, 1])
plt.show()
```

Setosa is having smaller feature and less distributed.
Versicolor is distributed in a average manner and average features
Virginica is highly distributed with large no .of values.

#### Violin plot 

```python
fig, axes = plt.subplots(2, 2, figsize=(16,10))
sns.violinplot( y="PW", x= "Classification", data=dataset, orient="v" , ax=axes[0, 0],inner='quartile')
sns.violinplot( y="PL", x= "Classification", data=dataset, orient="v" , ax=axes[0, 1],inner='quartile')
sns.violinplot( y="SL", x= "Classification", data=dataset, orient="v" , ax=axes[1, 0],inner='quartile')
sns.violinplot( y="SW", x= "Classification", data=dataset, orient="v" , ax=axes[1, 1],inner='quartile')
plt.show()
```

The above-plotted violin plot says that Iris-Setosa class is having a smaller petal length and petal width compared to other class.
Versicolor is distributed in a average manner and average features in case of petal length & width.
Virginica is highly distributed with large no .of values and features in case of sepal length & width.

#### Pair plot 

```python
sns.pairplot(dataset,hue='Classification')
plt.show
```


Setosa has both low petal length and width
Versicolor has both average petal length and width
Virginica has both high petal length and width.
Sepal width for setosa is high and length is low.
Versicolor have average values for for sepal dimensions.
Virginica has small width but large sepal length.
